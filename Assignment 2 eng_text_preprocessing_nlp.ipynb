```python
# 1.import all necessary modules
import nltk 
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import warnings
warnings.filterwarnings('ignore')
```

```python
# 2.download punkt_tab and add text
nltk.download('punkt_tab')
text = "I am happy to be coder and i love to gain knowledge which i don't no"
```

```python
# 3.tokenize text 
words = word_tokenize(text)
```

```python
# 4.clean and normalize tokenized text
cleaned_words = [word.lower() for word in words if word.isalpha()]
```

```python
# 5.Remove stop words 
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
filter_words = [word for word in cleaned_words if word not in stop_words]
```

```python
# 6.final output of text preprocessing
print('Original Text : \n', text)
print('\nTokenized Text : \n', words)
print('\n Filter Text : \n', ' '.join(filter_words))
```
