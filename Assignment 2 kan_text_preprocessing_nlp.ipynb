```python
# 1. import all necessary modules 
from indicnlp.tokenize import indic_tokenize
import re
```

```python
# 2. import text and do cleaning and normalization
text = "ಡೇಟಾ ಸೈನ್ಸ್‌ನ ಭಾಗವಾಗಿ ಮಷಿನ್ ಲರ್ನಿಂಗ್ ಮತ್ತು ಡೀಪ್ ಲರ್ನಿಂಗ್ ಪರಿಕಲ್ಪನೆಗಳ ಮೇಲೆ ಕೆಲಸ ಮಾಡುವುದನ್ನು ನಾನು ಆನಂದಿಸುತ್ತೇನೆ"
```

```python
text = 'ಸೈನ್ಸ್‌ನ ಭಾಗವಾಗಿ'
text = re.sub(r'‌ನ\b')
print(text)
```

```python
tokenized_text =list(indic_tokenize.trivial_tokenize(text))
print(tokenized_text)
```

```python
# 3. add stop words manually(predefined)
kannada_stop_words = {
    'ನಾನು', 'ಮತ್ತು', 'ಮೇಲೆ', 'ಭಾಗವಾಗಿ', 'ಮಾಡುವುದನ್ನು', 'ನ'
}
```

```python
def remove_suffix(word):
    # Remove zero width joiner
    word = word.replace("\u200c", "")
    
    # remove 'ನ' when it appears at end of the word
    word = re.sub(r"ನ$", "", word)
    return word
```

```python
# Apply suffix removal + stop word removal
filtered_words = []
for word in tokenized_text:
    cleaned = remove_suffix(word)
    if cleaned not in kannada_stop_words and cleaned.strip():
        filtered_words.append(cleaned)

print("Filtered:", filtered_words)
print("stop words:", len(tokenized_text) - len(filtered_words))
```

```python
# set do not all duplicate so instead of list[] we take set{}
word_freq = {word:filtered_words.count(word) for word in filtered_words}
print(word_freq)
```
